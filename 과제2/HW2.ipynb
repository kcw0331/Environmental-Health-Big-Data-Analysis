### **Q3 가격이 5000 이하인 다이아몬드들에 대해서, 다이아몬드 색에 따른 가격의 평균과 분산 정보를 Bar plot으로 그려보세요 (bar plot 의 x축은 색, y축은 색에따른 평균가격). 해당 bar plot에 표준편차를 errorbar 로 나타내세요. (Hint: ggplot에서 geom_errorbar 사용)**



```r
diamonds %>% group_by(color) %>% 
  #filter(price <= 5000) %>%
  summarise(price_mean = mean(price <= 5000, na.rm = TRUE), price_variance = var(price <= 5000, na.rm = TRUE), price_standards = sd(price <= 5000, na.rm = TRUE)) %>% head()
```

```r
diamondsna <- na.omit(diamonds) #처음에는 다이아몬드 데이터에서 비어있는 데이터는 제거를 하였습니다. 
diamonds1 <- diamondsna %>%  # 구해진 값을 diamonds1에 저장되도록 하였습니다.
  group_by(color) %>%
  summarise(mean = mean(price <= 5000), variance = var(price <= 5000) ,s.d = sd(price <= 5000), n = length(price <= 5000)) %>% #평균, 표주편차, 자료의 수
  mutate(s.e = s.d/sqrt(n)) #표준오차 
diamonds1
```

```r
library(ggplot2)
ggplot(diamonds1, aes(x = color, y = mean, color = color)) + 
  geom_col() + theme(text = element_text(size=25))
```

```r
library(ggplot2)
ggplot(diamonds1, aes(x = color, y = variance, color = color)) + 
  geom_col() + theme(text = element_text(size=25))
```

```r
library(ggplot2)
ggplot(diamonds1, aes(x = color, y = mean, color = color)) + 
  geom_errorbar(aes(x = color, ymin = mean - s.d, ymax = mean + s.e)) + theme(text = element_text(size=25))
```

* geom_errorbar만 사용하여 plot을 그려보았습니다.

```r
library(ggplot2)
ggplot(diamonds1, aes(x = color, y = mean, fill = color)) + 
  geom_bar(stat = 'identity') + 
  geom_errorbar(aes(x = color, ymin = mean - s.d, ymax = mean + s.e), width = 0.3, color = 'red') + theme(text = element_text(size=25))
```

* geom_bar와 bar 두개를 사용해서 plot을 그려보았습니다.
